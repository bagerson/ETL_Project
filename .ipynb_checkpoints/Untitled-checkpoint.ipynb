```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
premise_file = "Legally_Operating_Businesses.csv"
df = pd.read_csv(premise_file)
df.head()
```

```python
premise_file2 = "../NYPD_Complaint_Data_Historic.csv"
df2 = pd.read_csv(premise_file2)
df2.head()
```

```python
df.columns
```

```python
df2.columns
```

```python
reduced_df2 = df2[['CMPLNT_FR_DT',
                   'OFNS_DESC',
                   'LAW_CAT_CD',
                   'BORO_NM',
                   'PREM_TYP_DESC',
                   'SUSP_AGE_GROUP',
                   'SUSP_RACE',
                   'SUSP_SEX',
                   'Latitude',
                   'Longitude'
                ]].copy()

reduced_df2 = reduced_df2.rename(columns={'CMPLNT_FR_DT' : 'complaint_fr_date',
                                        'OFNS_DESC' : 'offense_desc',
                                        'LAW_CAT_CD' : 'law_cat_code',
                                        'BORO_NM' : 'boro_name',
                                        'PREM_TYP_DESC' : 'prem_type', 
                                        'SUSP_AGE_GROUP' : 'susp_age',
                                        'SUSP_RACE' : 'suspe_race',
                                        'SUSP_SEX' : 'susp_sex',
                                        'Latitude' : 'latitude',
                                        'Longitude' : 'longitude'
                                       })
reduced_df2.head()

```

```python
# reduced_df = df.loc[:, ["name", "goal", "pledged",
#                                     "state", "country", "staff_pick", "backers_count", "spotlight"]]
# reduced_df

reduced_df = df[['DCA License Number',
                 'License Expiration Date',
                 'License Status',
                 'License Creation Date',
                 'Industry', 
                 'Address Building',
                 'Address Street Name',
                 'Address ZIP',
                 'Address Borough',
                 'Borough Code',
                 'Latitude',
                 'Longitude'
                ]].copy()


reduced_df = reduced_df.rename(columns={'DCA License Number' : 'dca_license_number',
                                        'License Expiration Date' : 'license_expiration_date',
                                        'License Status' : 'license_status',
                                        'License Creation Date' : 'license_creation_date',
                                        'Industry' : 'industry', 
                                        'Address Building' : 'address_building',
                                        'Address Street Name' : 'address_street',
                                        'Address ZIP' : 'address_zip',
                                        'Address Borough' : 'address_borough',
                                        'Borough Code' : 'borough_code',
                                        'Latitude' : 'latitude',
                                        'Longitude' : 'longitude'
                                       })
reduced_df.head()
```

```python
connection_string = "postgres:@localhost:5432/ETLPROJECT_CRIME"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
reduced_df.to_sql(name='licenses', con=engine, if_exists='append', index=False)
# reduced_df.dtypes
```

```python
reduced_df2.to_sql(name='complaints', con=engine, if_exists='append', index=False)
```
