```python
#DCA license dataframe
#rename column headers to match SQL Schema and disply

reduced_df = df[['DCA License Number',
                 'License Expiration Date',
                 'License Status',
                 'License Creation Date',
                 'Industry', 
                 'Address Building',
                 'Address Street Name',
                 'Address ZIP',
                 'Address Borough',
                 'Borough Code',
                 'Latitude',
                 'Longitude'
                ]].copy()
```

```python
#DCA license dataframe
#rename column headers to match SQL Schema and disply
reduced_df = reduced_df.rename(columns={'DCA License Number' : 'dca_license_number',
                                        'License Expiration Date' : 'license_expiration_date',
                                        'License Status' : 'license_status',
                                        'License Creation Date' : 'license_creation_date',
                                        'Industry' : 'industry', 
                                        'Address Building' : 'address_building',
                                        'Address Street Name' : 'address_street',
                                        'Address ZIP' : 'address_zip',
                                        'Address Borough' : 'address_borough',
                                        'Borough Code' : 'borough_code',
                                        'Latitude' : 'latitude',
                                        'Longitude' : 'longitude'
                                       })
reduced_df.head()
```

```python
#connect to postgresql database
connection_string = "postgres:1234@localhost:5432/NYPDComplaint_Business"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#insert DCA license information into SQL database
reduced_df.to_sql(name='licenses', con=engine, if_exists='append', index=False)
# reduced_df.dtypes
```

```python
#insert NYPD complain information into SQL database

reduced_df2.to_sql(name='complaints', con=engine, if_exists='append', index=False)
```
